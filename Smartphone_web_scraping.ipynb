```python
import requests as rs 
from bs4 import BeautifulSoup as bs
import numpy as np
import pandas as pd
```

```python
page_num=1

links=[]
price=[]
model=[]
# parameters
internal_storage = []
ram = []
primary_camera = []
secondary_camera=[]
number_cores = []
processor = []

while page_num <=2:  # to access all pages on the web site (took only 2 page in terms of time)
    url = f'https://kontakt.az/telefonlar/mobil-telefonlar/page/{page_num}/'  
    r = rs.get(url)
    
    if r.status_code == 200:
        soup = bs(r.content, 'lxml')
        # to extract all smartphone links in the page
        for mob_link in soup.find_all('a', {'draggable':'false'}, href=True):
            links.append(str(mob_link.get('href'))) 
         
            # to parse phone links
            for html in links:
                r1 = rs.get(html) 
                soup1 = bs(r1.content, 'lxml')
            
                # to extract phone model
                model_ = soup1.find('div',{'class':"col-md-5"}).find('p').text
                model.append(model_)
            
                # to extract phone price
                try:
                    price_ = soup1.find('button', {'class':"buttonsvg feedback-action feaheart select-item"})['data-price']
                    price.append(price_)
                except:
                    price.append(np.nan)
            
                # to extract phone parameters
                try:
                    lis =soup1.find('div', {'class':"col-lg-8 col-md-6 col-sm-12 pl-0"},
                                {'class':"features-review-tabs submenuItems_params"})
                    parameters=[elements.text.strip('\n') for elements in lis.find_all('li')]

                    internal_storage.append(parameters[1])
                    ram.append(parameters[2])
                    primary_camera.append(parameters[3])
                    secondary_camera.append(parameters[4])
                    number_cores.append(parameters[5])
                    processor.append(parameters[9])
                                
                except:
                    internal_storage.append(np.nan)
                    ram.append(np.nan)
                    primary_camera.append(np.nan)
                    secondary_camera.append(np.nan)
                    number_cores.append(np.nan)
                    processor.append(np.nan)
                  
    else:
        break
                                    
    page_num+=1
```

```python
r.close()
```

```python
pd.set_option("max_columns",None)
pd.set_option("max_rows", None)
data = pd.DataFrame({
                      'Model':model,'Price':price, 'Internal Storage' : internal_storage,'Ram': ram,
                       'Primary Camera':primary_camera, 'Secondary Camera': secondary_camera,
                        'Number of Cores' : number_cores, 'Processor': processor
                       })
```
